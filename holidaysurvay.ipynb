```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = [1200, 850, 740, 590, 340,
        450, 890, 260, 610, 350,
        1780, 180, 850, 2050, 770,
        800, 1090, 510, 520, 220,
        1450, 280, 1120, 200, 350]
df = pd.DataFrame(data , columns=["Spending"])
print("Successfull Import")
```

```python
print("Minimum Spending: ", df["Spending"].min())
print("Maximum Spending: ", df["Spending"].max())
print("Average Spending: ", df["Spending"].mean())
print("Median Spending: ", df["Spending"].median())
print("Standard Deviation: ", df["Spending"].std())
```

```python
# Another fast Methode Using Describe
df.describe()
```

```python
bins = np.arange(0, 2251, 250)
df["Spending Range"] = pd.cut(df["Spending"], bins)

frequency = df["Spending Range"].value_counts().sort_index()
percent_frequency = (frequency / len(df)) * 100

print("Frequency Table:\n", frequency)
print("\nPercent Frequency:\n", percent_frequency)
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Spending"], bins=bins, edgecolor='black')
plt.title("Holiday Spending Distribution")
plt.xlabel("Spending Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Spending"])
plt.title("Boxplot of Holiday Spending")
plt.show()
```

```python
def segment(spend):
    if spend < 500:
        return "Low"
    elif spend <= 1000:
        return "Medium"
    else:
        return "High"

df["Segment"] = df["Spending"].apply(segment)

print(df["Segment"].value_counts())
```
